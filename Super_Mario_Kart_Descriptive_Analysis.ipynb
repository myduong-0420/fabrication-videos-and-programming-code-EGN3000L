# Super Mario Kart Descriptive analysis

Here, we gather 40 responses, thus 40 datapoints. When calculating the mean likelihood to buy this product, we found a score of 7.375/10. Twenty people contribute additional ideas for improvement.

```python
df.info()
```

```python
df.describe()
```

### Are there kids in the family?

```python
kids = df['Do you have kids or younger members on the family?'].value_counts()
```

```python
kids.plot(kind='pie', autopct=lambda p: '{:.2f}%({:.0f})'.format(p,(p/100)*kids.sum()), figsize=(5,5), title='Young member in family', xlabel='', ylabel='')
```

### Are they familiar with Super Mario's Kart?


```python
familiarity = df.iloc[:, 1].value_counts()
```

```python
familiarity.plot(kind='pie', autopct=lambda p: '{:.2f}%({:.0f})'.format(p,(p/100)*familiarity.sum()),  figsize=(5,5), title='Familiarity with product theme', xlabel='', ylabel='')
```

### How much are people willing to pay?

```python
price = df.iloc[:, 2].value_counts()
price.plot(kind='barh')
```

### What size and weight are preferable?

```python
size = df.iloc[:, 3].value_counts()
size.plot(kind='pie', autopct=lambda p: '{:.2f}%({:.0f})'.format(p,(p/100)*size.sum()),  figsize=(5,5), title='Product size', xlabel='', ylabel='')
```

```python
weight = df.iloc[:, 4].value_counts()
weight.plot(kind='pie', autopct=lambda p: '{:.2f}%({:.0f})'.format(p,(p/100)*weight.sum()),  figsize=(5,5), title='Product weight', xlabel='', ylabel='')
```

### Car noise

80% participants agree that we should include noises when designing the robot. This makes a lot of sense because it makes the robot more interesting, and safer for interaction with children since it may get in the way of children and trip them up.

```python
car_noise = df.iloc[:, 5].value_counts()
car_noise.plot(kind='pie', autopct=lambda p: '{:.2f}%({:.0f})'.format(p,(p/100)*car_noise.sum()),  figsize=(5,5), title='Car noise', xlabel='', ylabel='')
```

### Speed


```python
speed = df.iloc[:, -3].value_counts()
speed.plot(kind='pie', autopct=lambda p: '{:.2f}%({:.0f})'.format(p,(p/100)*speed.sum()),  figsize=(5,5), title='Recommended speed', xlabel='', ylabel='')
```
